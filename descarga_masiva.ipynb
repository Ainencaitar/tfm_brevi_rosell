```python
import requests
import csv
import io
import time

# Base URL for the API
base_url = "https://financialmodelingprep.com/api/v4/income-statement-bulk"

# API key for authentication
api_key = "NYR6U4zsGRBvbJUv4qZEF8h8A8MAmj91"

# Initialize a list to store all rows from each year
all_data = []

# Loop through each year from 1985 to 2023
for year in range(1985, 2024):
    # Construct the API URL with the current year and period
    url = f"{base_url}?year={year}&period=annual&apikey={api_key}"

    # Make the GET request to the API
    response = requests.get(url)

    # Check if the response status code is 200 (OK)
    if response.status_code == 200:
        # Convert the response text to a file-like object for csv.reader
        csv_data = io.StringIO(response.text)
        reader = csv.reader(csv_data)

        # Add each row of data to the all_data list
        for row in reader:
            all_data.append(row)
        print(f"Data for year {year} added.")
    else:
        print(f"Failed to fetch data for year {year}: {response.status_code} {response.text}")

    # Wait for 10 seconds before making the next request
    time.sleep(10)

# Export all collected data to a CSV file
csv_file_path = 'financial_data_1985_2023.csv'
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write all data rows
    writer.writerows(all_data)

print(f"Data exported to {csv_file_path}.")
```
